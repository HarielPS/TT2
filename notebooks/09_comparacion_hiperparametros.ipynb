```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
PROJECT_ROOT = Path("/home/harielpadillasanchez/Documentos/TT/TT2")
RULESET_DIR = PROJECT_ROOT / "outputs" / "ruleset_comparison"
OUT_DIR = PROJECT_ROOT / "outputs" / "comparison_finalists"
OUT_DIR.mkdir(parents=True, exist_ok=True)

MY_FILE = RULESET_DIR / "exp_ruleset_comparison_20260322_133408_summary_by_ruleset.csv"
PARTNER_FILE = RULESET_DIR / "exp_ruleset_comparison_20260404_031154_summary_by_ruleset.csv"

print("MY_FILE:", MY_FILE)
print("PARTNER_FILE:", PARTNER_FILE)
print("OUT_DIR:", OUT_DIR)
```

```python
df_me = pd.read_csv(MY_FILE)
df_partner = pd.read_csv(PARTNER_FILE)

df_me["owner"] = "hariel"
df_partner["owner"] = "nancy"

df_all = pd.concat([df_me, df_partner], ignore_index=True, sort=False)

print("Shape combinado:", df_all.shape)
display(df_all.head(3))
display(df_all.columns.tolist())
```

```python
optional_cols = [
    "gen_temperature",
    "gen_top_p",
    "gen_repetition_penalty",
    "gen_max_new_tokens",
]

for col in optional_cols:
    if col not in df_all.columns:
        df_all[col] = np.nan
print("Columnas normalizadas.")
```

```python
group_cols_preview = [
    c for c in [
        "owner",
        "model_key",
        "config_label",
        "ruleset",
        "gen_temperature",
        "gen_top_p",
        "gen_repetition_penalty",
        "gen_max_new_tokens",
        "sari",
        "bertscore_f1",
        "rougeL_f",
        "compression_ratio_eval",
        "exact_copy",
    ]
    if c in df_all.columns
]

display(df_all[group_cols_preview].head(10))
```

```python
df_rank = df_all.copy()

if "exact_copy" in df_rank.columns:
    df_rank = df_rank[df_rank["exact_copy"] < 0.50].copy()

if "compression_ratio_eval" in df_rank.columns:
    df_rank = df_rank[
        (df_rank["compression_ratio_eval"] > 0.60) &
        (df_rank["compression_ratio_eval"] < 1.60)
    ].copy()

df_rank["leader_score"] = (
    0.6 * df_rank["sari"] +
    0.4 * (df_rank["bertscore_f1"] * 100.0)
)

print("Shape después de filtros:", df_rank.shape)
display(
    df_rank[
        [c for c in [
            "owner", "model_key", "config_label", "ruleset",
            "sari", "bertscore_f1", "leader_score",
            "compression_ratio_eval", "exact_copy"
        ] if c in df_rank.columns]
    ].head(10)
)
```

```python
def safe_str(v):
    if pd.isna(v):
        return "NA"
    return str(v)

df_rank["config_signature"] = df_rank.apply(
    lambda r: " | ".join([
        safe_str(r.get("owner")),
        safe_str(r.get("model_key")),
        safe_str(r.get("config_label")),
        safe_str(r.get("ruleset")),
        f"temp={safe_str(r.get('gen_temperature'))}",
        f"top_p={safe_str(r.get('gen_top_p'))}",
        f"rep={safe_str(r.get('gen_repetition_penalty'))}",
        f"max_new={safe_str(r.get('gen_max_new_tokens'))}",
    ]),
    axis=1
)

display(df_rank[["config_signature"]].head(5))
```

```python
ranking_global = df_rank.sort_values(
    by=["leader_score", "sari", "bertscore_f1"],
    ascending=False
).reset_index(drop=True)

display(
    ranking_global[
        [c for c in [
            "owner",
            "model_key",
            "config_label",
            "ruleset",
            "gen_temperature",
            "gen_top_p",
            "gen_repetition_penalty",
            "gen_max_new_tokens",
            "sari",
            "bertscore_f1",
            "rougeL_f",
            "compression_ratio_eval",
            "exact_copy",
            "leader_score",
        ] if c in ranking_global.columns]
    ].head(20)
)

ranking_global.to_csv(OUT_DIR / "ranking_global.csv", index=False, encoding="utf-8-sig")
print(OUT_DIR / "ranking_global.csv")
```

```python
top5_by_model = (
    df_rank
    .sort_values(by=["model_key", "leader_score", "sari", "bertscore_f1"], ascending=[True, False, False, False])
    .groupby("model_key", as_index=False, group_keys=False)
    .head(5)
    .reset_index(drop=True)
)

display(
    top5_by_model[
        [c for c in [
            "owner",
            "model_key",
            "config_label",
            "ruleset",
            "gen_temperature",
            "gen_top_p",
            "gen_repetition_penalty",
            "gen_max_new_tokens",
            # "gen_do_sample",
            # "gen_no_repeat_ngram_size",
            "sari",
            "bertscore_f1",
            "leader_score",
        ] if c in top5_by_model.columns]
    ]
)

top5_by_model.to_csv(OUT_DIR / "top5_by_model.csv", index=False, encoding="utf-8-sig")
print(OUT_DIR / "top5_by_model.csv")
```

```python
top5_by_owner = (
    df_rank
    .sort_values(by=["owner", "leader_score", "sari", "bertscore_f1"], ascending=[True, False, False, False])
    .groupby("owner", as_index=False, group_keys=False)
    .head(5)
    .reset_index(drop=True)
)

display(
    top5_by_owner[
        [c for c in [
            "owner",
            "model_key",
            "config_label",
            "ruleset",
            "gen_temperature",
            "gen_top_p",
            "gen_repetition_penalty",
            "gen_max_new_tokens",
            "sari",
            "bertscore_f1",
            "leader_score",
        ] if c in top5_by_owner.columns]
    ]
)

top5_by_owner.to_csv(OUT_DIR / "top5_by_owner.csv", index=False, encoding="utf-8-sig")
print(OUT_DIR / "top5_by_owner.csv")
```

```python
top3_by_model_owner = (
    df_rank
    .sort_values(
        by=["owner", "model_key", "leader_score", "sari", "bertscore_f1"],
        ascending=[True, True, False, False, False]
    )
    .groupby(["owner", "model_key"], as_index=False, group_keys=False)
    .head(3)
    .reset_index(drop=True)
)

display(
    top3_by_model_owner[
        [c for c in [
            "owner",
            "model_key",
            "config_label",
            "ruleset",
            "gen_temperature",
            "gen_top_p",
            "gen_repetition_penalty",
            "gen_max_new_tokens",
            "sari",
            "bertscore_f1",
            "leader_score",
        ] if c in top3_by_model_owner.columns]
    ]
)

top3_by_model_owner.to_csv(OUT_DIR / "top3_by_model_owner.csv", index=False, encoding="utf-8-sig")
print(OUT_DIR / "top3_by_model_owner.csv")
```

```python
print("Resumen rapido")
print("-" * 50)
print("Total configs comparadas:", len(df_rank))
print("Owners:", df_rank["owner"].unique().tolist())
print("Modelos:", df_rank["model_key"].unique().tolist())

best_global = ranking_global.iloc[0]
print("\nMejor config global:")
for col in [
    "owner", "model_key", "config_label", "ruleset",
    "gen_temperature", "gen_top_p", "gen_repetition_penalty",
    "gen_max_new_tokens",
    "sari", "bertscore_f1", "leader_score"
]:
    if col in best_global.index:
        print(f"{col}: {best_global[col]}")
```
